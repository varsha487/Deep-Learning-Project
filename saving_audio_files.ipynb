```python
import librosa
import soundfile as sf
from IPython.display import Audio

guitar_file = 'let_it_go_guitar.wav'
piano_file = 'let_it_go_piano.wav'

y_guitar, sr_g = librosa.load(guitar_file, sr=None)  # sr=None preserves original sample rate
y_piano, sr_p = librosa.load(piano_file, sr=None)

print(f"Guitar length: {len(y_guitar)/sr_g:.2f} sec")
print(f"Piano length: {len(y_piano)/sr_p:.2f} sec")
```

```python
y, sr = librosa.load('let_it_go_piano.wav', sr=None)
start_sec = 2.6 
end_sec = 20  
Audio(y[int(start_sec*sr):int(end_sec*sr)], rate=sr)
```

```python
start_sec = 2.4 

# Convert to sample indices
start_sample = int(start_sec * sr)

y_trimmed = y[start_sample:]

# Save trimmed file
import soundfile as sf
sf.write('let_it_go_pianot.wav', y_trimmed, sr)
```
