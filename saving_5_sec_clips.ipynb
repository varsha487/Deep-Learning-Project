```python
import librosa
import soundfile as sf
from IPython.display import Audio, display
import os
```

```python
#  Load trimmed audio 
guitar_file = 'dm_guitar.wav'
piano_file = 'dm_piano.wav'
```

```python
y_guitar, sr_g = librosa.load(guitar_file, sr=None)
y_piano, sr_p = librosa.load(piano_file, sr=None)

# Ensure sample rates match
if sr_g != sr_p:
    y_piano = librosa.resample(y_piano, orig_sr=sr_p, target_sr=sr_g)
    sr_p = sr_g
    
# Create dataset folders 
os.makedirs('dataset/guitar', exist_ok=True)
os.makedirs('dataset/piano', exist_ok=True)
```

```python
# seconds to test**
guitar_start_sec = 55
piano_start_sec  = 56

chunk_len_sec = 5.0

g_start = int(guitar_start_sec * sr_g)
g_end   = g_start + int(chunk_len_sec * sr_g)
p_start = int(piano_start_sec * sr_p)
p_end   = p_start + int(chunk_len_sec * sr_p)

guitar_preview = y_guitar[g_start:g_end]
piano_preview  = y_piano[p_start:p_end]
```

```python
g_preview = guitar_preview.copy()
p_preview = piano_preview.copy()

print(f"Guitar preview: {guitar_start_sec}â€“{guitar_start_sec+chunk_len_sec}")
display(Audio(data=g_preview, rate=sr_g))

print(f"Piano preview: {piano_start_sec}â€“{piano_start_sec+chunk_len_sec}")
display(Audio(data=p_preview, rate=sr_p))
```

```python
print(f"Guitar preview: {guitar_start_sec}â€“{guitar_start_sec+chunk_len_sec}")
display(Audio(guitar_preview, rate=sr_g))
```

```python
print(f"Piano preview: {piano_start_sec}â€“{piano_start_sec+chunk_len_sec}")
Audio(piano_preview, rate=sr_p)
```

```python
clip_id = 31  # ***change clip # !!!

sf.write(f"dataset/guitar/clip_{clip_id}.wav", guitar_preview, sr_g)
sf.write(f"dataset/piano/clip_{clip_id}.wav", piano_preview, sr_p)

print(f"Saved clip_{clip_id} for both guitar & piano!")
```
